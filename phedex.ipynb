```python
%matplotlib inline
%pylab inline
import numpy as np
import pylab
import pandas as pd 
from glob import glob 
import json
import hashlib

pylab.rcParams['figure.figsize'] = (8.0, 6)
import matplotlib.pyplot as plt
from sklearn import datasets
from time import time 
from IPython.display import display, HTML
```

```python
# load the data 

store = pd.HDFStore('data/errors.h5')
if True:
    now = time()
    data = pd.DataFrame() # master dataframe 
    for f in glob('data/jsons/*json'):
        try:
#            print('Loading '+f)
            payload = json.load(open(f))['phedex']['link']
            tmp_df = pd.DataFrame([
                            {'from':link['from_id'],
                             'to':link['to_id'],
                             'from_site':link['from'],
                             'to_site':link['to'],
                             'file':f['name'],
                             'f_hash':int(hashlib.sha256(
                                                  f['name'].encode('utf-8')
                                              ).hexdigest(), 16) % 10**8,
                             'size':f['size'] / 1e9,
                             'time':now - e['time_xfer'],
                             'e':str(e['detail_log']['$t']).replace(f['name'],'X.root'),
                             'e_hash':int(hashlib.sha256(
                                                  str(e['detail_log']['$t']).replace(f['name'],'X.root').encode('utf-8')
                                              ).hexdigest(), 16) % 10**8}
                            for link in payload
                            for b in link['block']
                            for f in b['file']
                            for e in f['transfer_error']
                        ])
            data = data.append(tmp_df, ignore_index=True)
        except json.JSONDecodeError as e:
            print(e)
    store['errors'] = data 
else:
    data = store['errors']
store.close()
data.head()
```

```python
# try some clustering 

from sklearn.cluster import DBSCAN, KMeans
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from kmodes.kprototypes import KPrototypes

scaler = StandardScaler()
X = scaler.fit_transform(data.as_matrix(columns=['e_hash','f_hash','from','to','size','time']))
N_CLUSTERS = 10
model = KPrototypes(n_clusters=N_CLUSTERS)
predictions = model.fit_predict(X, categorical=[0,1,2,3])
```

```python
#predictions = model.predict(X)
idx = np.arange(X.shape[0])

for i in range(N_CLUSTERS):
    print('Cluster {}'.format(i))
    mask = predictions==i
    idx_ = idx[mask]
    display(data.loc[idx_.tolist(),:])
```
